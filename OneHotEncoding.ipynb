# OneHot Encoding with Top Categories

```python
counts = df['brand'].value_counts()
df['brand'].nunique()
threshold=100
rep=counts[counts<= threshold].index #taking those car<100
```

```python
pd.get_dummies(df['brand'].replace(rep,'uncommon'))
```
